```python
import numpy as np

A = np.array([1,2,3])
B = np.array([4,5,6])

print(A + B)
```

```python
arr1 = np.array([2,4,6])
arr2 = np.array([1,2,3])

print(arr1 * arr2)
```

```python
arr1 = np.array([10,20,30])
arr2 = np.array([2,4,5])

print(arr1 / arr2)
```

```python
arr = np.array([4,9,16,25])
print(np.sqrt(arr))
```

```python
arr = np.array([2,3,4])
print(arr ** 3)
```

```python
arr = np.array([5,10,15,20])

print(arr.sum())
print(arr.mean())
print(arr.max())
print(arr.min())
```

```python
A = np.array([1,2,3])
B = np.array([4,5,6])

print(np.dot(A, B))
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(np.matmul(A, B))
```

```python
arr = np.array([1,2,3,4,5])

print(np.var(arr))   # variance
print(np.std(arr))   # standard deviation
```

```python
arr = np.array([1,2,3,4])
print(np.cumsum(arr))
```

```python
arr = np.array([1,2,3,4])
print(np.cumprod(arr))
```

```python
arr = np.array([2.3456, 7.8911, 3.4567])
print(np.round(arr, 2))
```
