```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

plt.plot(x, y)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ["0", "π/2", "π", "3π/2", "2π"])
plt.title("Sine Wave with Custom Ticks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)
y = np.exp(-x)

plt.plot(x, y)
plt.yscale("log")
plt.title("Exponential Decay (Log Scale)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 200)
y = x**3 - x

plt.figure(figsize=(10, 4))
plt.plot(x, y)
plt.title("y = x^3 - x")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), linestyle="-", label="sin(x)")
plt.plot(x, np.cos(x), linestyle="--", label="cos(x)")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, np.sin(x))
axes[0].set_title("sin(x)")

axes[1].plot(x, np.cos(x))
axes[1].set_title("cos(x)")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

plt.plot(x, y)
plt.annotate("Max = 1", xy=(np.pi/2, 1),
             xytext=(2, 1.2),
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 300)
y = np.tan(x)

plt.plot(x, y)
plt.ylim(-5, 5)
plt.title("Tan(x) with Limited Y-axis")
plt.show()
```
