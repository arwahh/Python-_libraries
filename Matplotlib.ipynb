```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 500)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1 + y2

fig, axes = plt.subplots(3, 1, sharex=True, figsize=(6, 8))

axes[0].plot(x, y1)
axes[0].set_ylabel("sin(x)")
axes[0].set_title("Sine, Cosine and Sum")

axes[1].plot(x, y2)
axes[1].set_ylabel("cos(x)")

axes[2].plot(x, y3)
axes[2].set_ylabel("sin(x) + cos(x)")
axes[2].set_xlabel("x")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]
temp = [30, 32, 31, 29, 28, 27, 26]
humidity = [40, 45, 50, 60, 65, 70, 75]

fig, ax1 = plt.subplots()

ax1.plot(days, temp, marker="o")
ax1.set_xlabel("Day")
ax1.set_ylabel("Temperature (C)")

ax2 = ax1.twinx()
ax2.plot(days, humidity, marker="s")
ax2.set_ylabel("Humidity (%)")

plt.title("Temperature and Humidity Over a Week")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x)

plt.plot(x, y)

ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
labels = ["0", "/2", "", "3/2", "2"]

plt.xticks(ticks, labels)
plt.yticks([-1, -0.5, 0, 0.5, 1])

plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Wave with Custom Ticks")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

x = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(8, 6))
gs = GridSpec(2, 2, figure=fig)

ax_top = fig.add_subplot(gs[0, :])      # full top row
ax_bottom_left = fig.add_subplot(gs[1, 0])
ax_bottom_right = fig.add_subplot(gs[1, 1])

ax_top.plot(x, np.sin(x))
ax_top.set_title("Top: sin(x)")

ax_bottom_left.plot(x, np.cos(x))
ax_bottom_left.set_title("Bottom Left: cos(x)")

ax_bottom_right.plot(x, np.tan(x))
ax_bottom_right.set_ylim(-5, 5)
ax_bottom_right.set_title("Bottom Right: tan(x)")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)

plt.plot(x, x, label="x")
plt.plot(x, x**2, label="x^2")
plt.plot(x, x**3, label="x^3")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Polynomials")

plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

plt.imshow(Z, extent=[x.min(), x.max(), y.min(), y.max()],
           origin="lower", aspect="auto")
plt.colorbar(label="Value")
plt.title("sin(X) * cos(Y)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 200)
y = x**2

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("x^2")
plt.title("Parabola with Minimum Annotated")

min_x = 0
min_y = 0

plt.scatter(min_x, min_y)
plt.annotate("Minimum (0, 0)",
             xy=(min_x, min_y),
             xytext=(1, 10),
             textcoords="data",
             arrowprops=dict(arrowstyle="->"))

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y = 10**x

plt.plot(x, y)
plt.yscale("log")

plt.xlabel("x")
plt.ylabel("10^x (log scale)")
plt.title("Exponential on Logarithmic Scale")
plt.grid(True, which="both")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = (8, 4)
plt.rcParams["font.size"] = 12
plt.rcParams["axes.grid"] = True

x = np.linspace(0, 2 * np.pi, 200)

plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine and Cosine with Global Settings")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)
y = np.exp(-x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("exp(-x)")
plt.title("Exponential Decay")
plt.tight_layout()

plt.savefig("decay.png", dpi=300)
plt.show()
```
