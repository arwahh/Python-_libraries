```python
import numpy as np

lst = [3, 6, 9, 12]
arr = np.array(lst)
print(arr)
```

```python
import numpy as np

arr = np.arange(2, 22, 2)
print(arr)
```

```python
arr = np.array([5, 10, 15, 20, 25])
print(arr[2])
```

```python
arr = np.array([10,20,30,40,50,60])
print(arr[2:5])
```

```python
arr = np.array([1,2,3,4,5])
rev = arr[::-1]
print(rev)
```

```python
import numpy as np

arr = np.arange(1, 10).reshape(3, 3)
print(arr)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr.shape)  # (2, 3)
print(arr.size)   # 6 elements
print(arr.dtype)  # int64 or int32
```

```python
import numpy as np

A = np.array([1,2,3])
B = np.array([4,5,6])

print(A + B)
```

```python
import numpy as np

arr = np.array([[1,2],[3,4]])
print(arr * 5)
```
