```python
#adding columns
import pandas as pd
data={
    "Name":["arwah","seher","falak","simran","arooj","khusbhoo"],
    "Age":[22,25,28,23,23,22],
    "Salary":[80000,50000,90000,45000,50000,60000],
    "Performance_score":[90,95,80,50,88,80]
}
df=pd.DataFrame(data)
print(df)
df["bonus"]=df["Salary"]*0.1
print(df)
```
