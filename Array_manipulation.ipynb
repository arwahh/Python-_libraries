```python

import numpy as np

arr = np.array([1,2,3,4,5,6])
reshaped = arr.reshape(2, 3)
print(reshaped)
```

```python
arr = np.array([[1,2,3],[4,5,6]])
flat = arr.flatten()
print(flat)
```

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print(arr.T)
```

```python
A = np.array([1,2,3])
B = np.array([4,5,6])

print("Horizontal:", np.hstack([A, B]))
print("Vertical:\n", np.vstack([A, B]))
```

```python
arr = np.array([10,20,30,40])
print(np.split(arr, 2))
```

```python
arr = np.array([1,2,3,4])
column = arr[:, np.newaxis]
print(column)
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(np.dstack((A, B)))
```

```python
arr = np.array([1,2,3])
print(np.repeat(arr, 3))
```

```python
print(np.tile(np.array([1,2]), 3))
```

```python
arr = np.array([[1,2],[3,4]])
swapped = arr[[1, 0]]
print(swapped)
```

```python
arr = np.array([[10,20,30],[40,50,60]])
new_arr = np.delete(arr, 1, axis=1)
print(new_arr)
```
