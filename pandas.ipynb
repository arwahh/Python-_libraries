```python
import pandas as pd

data = {
    "Name": ["Aman", "Riya", "Sohan"],
    "Age": [21, 22, 20],
    "City": ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["Age"]
```

```python
df[["Name", "City"]]
```

```python
df[df["Age"] > 21]
```

```python
df["Passed"] = ["Yes", "Yes", "No"]
print(df)
```

```python
df["Age"] = df["Age"] + 1
```

```python
df.drop("City", axis=1, inplace=True)
```

```python
df.sort_values(by="Age")
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
data = {
    "Dept": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 60000, 45000]
}

df = pd.DataFrame(data)

df.groupby("Dept")["Salary"].mean()
```

```python
df.rename(columns={"Salary": "Monthly_Salary"}, inplace=True)
```
