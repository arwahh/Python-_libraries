```python
import numpy as np

arr = np.array([2, 4, np.nan, 8, np.nan, 12])
print(np.isnan(arr))
```

```python
import numpy as np

arr = np.array([2, 4, np.nan, 8, np.nan, 12])
print(np.sum(np.isnan(arr)))
```

```python
import numpy as np

arr = np.array([2, 4, np.nan, 8, np.nan, 12])
arr_filled = np.nan_to_num(arr, nan=0)

print(arr_filled)
```

```python
import numpy as np

arr = np.array([2, 4, np.nan, 8, np.nan, 12])
mean_val = np.nanmean(arr)     # mean ignoring NaN

arr2 = np.where(np.isnan(arr), mean_val, arr)

print("Mean =", mean_val)
print(arr2)
```

```python
import numpy as np

arr = np.array([5, np.nan, 9, np.nan, 12])
clean = arr[~np.isnan(arr)]

print(clean)
```

```python
import numpy as np

arr = np.array([[1, 2, np.nan],
                [4, 5, 6],
                [np.nan, 8, 9]])

print(np.isnan(arr).any(axis=1))
```

```python
import numpy as np

arr = np.array([[1, np.nan, 3],
                [4, 5, np.nan],
                [np.nan, 8, 9]])

row_means = np.nanmean(arr, axis=1)

# Replace NaN using broadcasting
inds = np.where(np.isnan(arr))
arr[inds] = np.take(row_means, inds[0])

print(arr)
```

```python
import numpy as np

arr = np.array([[1, np.nan, 3],
                [4, 5, np.nan],
                [np.nan, 8, 9]])

col_means = np.nanmean(arr, axis=0)
inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_means, inds[1])

print(arr)
```
