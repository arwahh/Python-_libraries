```python
import random

num = random.randint(1, 10)
print("Random number between 1 and 10:", num)
```

```python
import random

x = random.random()   # 0.0 <= x < 1.0
print("Random float:", x)
```

```python
import random

x = random.uniform(5, 10)   # 5.0 <= x <= 10.0
print("Random float between 5 and 10:", x)
```

```python
import random

colors = ["red", "green", "blue", "yellow", "black"]
choice = random.choice(colors)
print("Random color:", choice)
```

```python
import random

nums = list(range(1, 11))
print("Before shuffle:", nums)

random.shuffle(nums)
print("After shuffle:", nums)
```

```python
import random

otp = ""
for _ in range(6):
    digit = random.randint(0, 9)
    otp += str(digit)

print("Your OTP is:", otp)
```

```python
import random

results = []
for _ in range(10):
    roll = random.randint(1, 6)
    results.append(roll)

print("Dice rolls:", results)
```

```python
import random

numbers = list(range(1, 51))
ticket = random.sample(numbers, 6)   # returns 6 unique elements

print("Your lottery numbers are:", ticket)
```

```python
import random
import string  # built-in module for characters

characters = string.ascii_letters + string.digits
password = "".join(random.choice(characters) for _ in range(8))

print("Random password:", password)
```
