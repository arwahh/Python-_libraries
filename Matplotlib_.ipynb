```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = 2 * x + 1

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot: y = 2x + 1")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y1 = x
y2 = x**2

plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = x^2")
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
import matplotlib.pyplot as plt

names = ["A", "B", "C", "D"]
marks = [70, 90, 50, 80]

plt.bar(names, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students Marks (Bar Chart)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title("Histogram of Random Numbers")
plt.show()
```

```python
import matplotlib.pyplot as plt

companies = ["A", "B", "C"]
share = [40, 35, 25]

plt.pie(share, labels=companies, autopct="%1.1f%%")
plt.title("Market Share")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.title("Sine Wave")

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.title("Cosine Wave")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10)
y = 3 * x

plt.plot(x, y, linestyle="--", color="red")
plt.title("Dashed Red Line")
plt.show()
```
